```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import glob
from scipy.stats import linregress
from matplotlib.colors import LightSource
from matplotlib import cm
import pickle
#from landlab import RasterModelGrid, imshow_grid
import matplotlib as mpl
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from scipy.signal import savgol_filter
from scipy.stats import linregress
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.gridspec import GridSpec
```

```python
#Plot output from an example simulation
fz = "time-dependent-experiments_05m_2_z_trans.csv"
fs = "time-dependent-experiments_05m_2_s_trans.csv"
z = pd.read_csv(fz)
s = pd.read_csv(fs)
z = z.astype(float)
s = s.astype(float)

s = s[(s['x'] >= 1.) & (s['x'] <= 4.)]
z = z[(z['x'] >= 1.) & (z['x'] <= 4.)]

cols = list(z)
ls = ["dashed", "dotted", "dashdot", "solid"]

fig = plt.figure(figsize = (4., 12.))
gs = GridSpec(5, 1, figure=fig, height_ratios=[1,1,1,1,1])#,2])


count = 0
for i, j in enumerate(cols):
    if j == '0' or j == '12.0' or j == '24.0' or j == '36.0':

        ########Transects plot
        axn = fig.add_subplot(gs[count+1])
        if j == '0':
            axn.annotate("(a)", xy = (-0.35, 1.15), xycoords = 'axes fraction', fontsize = 12)
        axn.fill_between(s[cols[0]]-2.5, s[cols[i]]-100, z[cols[i]]-100, color = 'lightgray', zorder = 2)
        axn.plot(z[cols[0]]-2.5, z[cols[i]]-100, c = 'k', linewidth = 1)
        axn.plot(s[cols[0]]-2.5, s[cols[i]]-100, c = 'steelblue')
        axn.annotate(" {} days".format(int(float(cols[i]))), xy = (1.26,z[cols[i]].min()-100+0.05), fontsize = 8)
        axn.set_aspect('equal')
        axn.set_ylim([s[cols[i]].min()-100.1, s[cols[i]].min()-99.4])
        axn.set_ylabel("$\\Delta z$ (m)", fontsize = 11)
        if count != 3:
            axn.set_xticklabels([])
            axn.set_xticks([-1.5, -1.0, -0.5, 0, .5, 1.0, 1.5])
        if count == 3:
            axn.set_xlabel('Postion (m)', fontsize = 11)
            axn.set_xticks([-1.5, -1.0, -0.5, 0, .5, 1.0, 1.5])
        count += 1


plt.subplots_adjust(hspace=-1.08)
plt.tight_layout()
sns.despine()
plt.savefig('cross-sections.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
#Plot output from an example simulation

fz = "time-dependent-experiments_05m_2_z_trans.csv"
fs = "time-dependent-experiments_05m_2_s_trans.csv"
z = pd.read_csv(fz)
s = pd.read_csv(fs)
z = z.astype(float)
s = s.astype(float)

s = s[(s['x'] >= 1.5) & (s['x'] <= 3.5)]
z = z[(z['x'] >= 1.5) & (z['x'] <= 3.5)]

cols = list(z)
ls = ["dashed", "dotted", "dashdot", "solid"]

fig = plt.figure(figsize = (7,5))

count = 0
for i, j in enumerate(cols):
    if j == '24.0' :

        ########Transects plot
        axn = fig.add_subplot()
        if j == '0':
            axn.annotate("(a)", xy = (-0.35, 1.15), xycoords = 'axes fraction', fontsize = 12)
        axn.fill_between(s[cols[0]]-2.5, s[cols[i]]-100, z[cols[i]]-100, color = 'lightgray', zorder = 2)
        axn.plot(z[cols[0]]-2.5, z[cols[i]]-100, c = 'k', linewidth = 1, label = "$z$")
        axn.plot(s[cols[0]]-2.5, s[cols[i]]-100, c = 'steelblue', label = "$s$")
        axn.set_ylim([s[cols[i]].min()-100.2, s[cols[i]].min()-99.4])
        axn.set_ylabel("Elevation (m)", fontsize = 11)
        axn.set_xlabel("$x$ (m)", fontsize = 11)
        axn.set_aspect('equal')
        count += 1

axn.set_xticklabels([])
axn.set_yticklabels([])
plt.legend(loc = 'upper left')
plt.subplots_adjust(hspace=-1.08)
plt.tight_layout()
sns.despine()
#plt.savefig('diagram1.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
#Plot output from an example simulation
fz = "time-dependent-experiments_05m_2_z_trans.csv"
fs = "time-dependent-experiments_05m_2_s_trans.csv"
z = pd.read_csv(fz)
s = pd.read_csv(fs)
z = z.astype(float)
s = s.astype(float)

s = s[(s['x'] >= 2.5) & (s['x'] <= 4)]
z = z[(z['x'] >= 2.5) & (z['x'] <= 4)]

cols = list(z)
#ls = ["solid", "dashed", "dotted", "dashdot"]

ls = ["solid", "dashed", "dotted", "dashdot"]

fig, ax = plt.subplots(2,1, figsize = (4., 5.), sharex = True)

colors = plt.cm.cool_r(np.linspace(0, 1, 20))
labels = ['0 days', '12 days', '24 days', '36 days']

count = 0
for i, j in enumerate(cols):
    #if j == '0' or j == '6.0'or j == '12.0' or j == '24.0' or j == '30.0' or j == '36.0':
    #if j != 'x' and float(j) % 2 == 0:
    if j == '0' or j == '12.0' or j == '24.0' or j == '36.0':
        color = colors[count % len(colors)]
        
        ax[0].plot(z[cols[0]]-2.5, z[cols[i]]-s[cols[i]], c = 'k', linewidth = 1.2, linestyle = ls[count], label = labels[count])
        if count == 0:
            ax[0].annotate(" {} days".format(int(float(cols[i]))), xy = (-0.22, (z[cols[i]]-s[cols[i]]).max()+0.01), fontsize = 8)
        else:
            ax[0].annotate(" {} days".format(int(float(cols[i]))), xy = (-0.23, (z[cols[i]]-s[cols[i]]).max()+0.01), fontsize = 8)

        ax[1].plot(z[cols[0]]-2.5, savgol_filter(z[cols[i]], window_length = 5, polyorder = 2, deriv = 1, delta = 0.025) , linestyle = ls[count], color = 'k', label = labels[count])

        count += 1
ax[1].set_xlabel("Position (m)", fontsize = 11)
ax[1].set_ylabel('Slope (m/m)', fontsize = 11)
ax[0].set_ylabel('Debris Thickness (m)', fontsize = 11)
ax[0].annotate("(b)", xy = (-0.2, 1.01), xycoords = 'axes fraction', fontsize = 12)
ax[1].annotate("(c)", xy = (-0.2, 1.01), xycoords = 'axes fraction', fontsize = 12)
ax[0].legend()
sns.despine()
plt.tight_layout()
plt.savefig('deb-slope-transects.png', dpi = 300)
plt.show()
```
